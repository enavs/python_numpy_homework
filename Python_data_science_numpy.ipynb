### Enrollments Data

```python
enrollments = pd.read_csv('../day_2/enrollments.csv')    
enrollments
```

```python
enrollments.dtypes
```

```python

column_names = ["account_key", "status", "join_date", "cancel_date", "days_to_cancel", "is_enrolled", "is_canceled"]
enrollments = enrollments.reindex(columns=column_names)
enrollments
```

```python
# enrollments.groupby("assigned_rating").account_key.agg(['min', 'max', 'sum', 'count'])
status_count = enrollments.groupby("status").account_key.agg(['count'])
status_count
```

```python
print(f'There are {len(enrollments)} records in the enrollment file')
print(f'There are {len(status_count)} different types of statuses in the enrollment file')
```

```python
status_enrolled_count = enrollments.groupby("status").agg({'account_key': ['count'], 'days_to_cancel': ['sum', 'min','max','mean']})
status_enrolled_count
```

```python
status_enrolled_count = enrollments.groupby("status").agg(
    record_count=pd.NamedAgg(column="account_key", aggfunc="count"),
    days_to_cancel_sum=pd.NamedAgg(column="days_to_cancel", aggfunc="sum"))
status_enrolled_count
```

### Engagement Data

```python
engagements = pd.read_csv('../day_2/engagements.csv')    
engagements
```

```python
engagements.dtypes
```

```python
engagements['account_key'] = engagements['acct']

column_names = ["account_key", "utc_date", "num_courses_visited", "total_minutes_visited", "lessons_completed", "projects_completed"]
engagements = engagements.reindex(columns=column_names)
engagements
```

```python
# engagements.groupby("assigned_rating").account_key.agg(['min', 'max', 'sum', 'count'])
engagement_count = engagements.groupby("num_courses_visited").account_key.agg(['count'])
engagement_count
```

```python
print(f'There are {len(engagements)} records in the engagement file')
print(f'There are {len(engagement_count)} different number of courses visited in the engagement file')
```

```python
engagement_count2 = engagements.groupby("num_courses_visited").agg({'account_key': ['count'], 'total_minutes_visited': ['sum', 'min','max','mean'], 'lessons_completed': ['sum', 'min','max','mean'], 'projects_completed': ['sum', 'min','max','mean']})
engagement_count2
```

## Question 1: How many people showed no engagement?
### This is defined where num_courses_visited = 0

```python
engagement_per_acct = engagements.groupby("account_key").agg(
    num_courses_visited_sum=pd.NamedAgg(column="num_courses_visited", aggfunc="sum"),
    total_minutes_visited_sum=pd.NamedAgg(column="total_minutes_visited", aggfunc="sum"),
    lessons_completed_sum=pd.NamedAgg(column="lessons_completed", aggfunc="sum"),
    projects_completed_sum=pd.NamedAgg(column="projects_completed", aggfunc="sum")
)
engagement_per_acct
```

```python
engagement_per_acct.dtypes
```

```python
engagement_per_acct.describe()
```

```python
engagement_per_acct_answer1 = engagement_per_acct[(engagement_per_acct.num_courses_visited_sum == 0)]
engagement_per_acct_answer1
```

### Answer to question 1

```python
print(f"There are {len(engagement_per_acct_answer1)} people who showed zero engagement")
```

## Question 2: What is the count of those who have completed 1 or more lessons?
### This is defined where lessons_completed >= 1

```python
lesson_per_acct_answer2 = engagement_per_acct[(engagement_per_acct.lessons_completed_sum > 1)]
lesson_per_acct_answer2
```

### Answer to question 2

```python
print(f"There are {len(lesson_per_acct_answer2)} people who showed zero engagement")
```
